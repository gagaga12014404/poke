```python
import requests
import bs4
from bs4 import BeautifulSoup
import json
```

```python
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
def GetSelect(url, sel):
    r = requests.get(url, verify=False)
    soup = BeautifulSoup(r.text, "html.parser")
    return soup.select(sel)
def GetFindAll(url,find):
    r = requests.get(url, verify=False)
    soup = BeautifulSoup(r.text, "html.parser")
    return soup.find_all(find)
def GetHp(url):
    hp=GetSelect(url,"td")
    hpList=[]
    for i in hp:
        hpList.append(str(i))
    for i in hpList:
        if "/255" in i:
            at = i.find("/255")
            print(i[at-3:at])
            yield i[at-3:at]
```

```python
main_url = "https://wiki.52poke.com/wiki"
national_list = main_url+"/%E5%AE%9D%E5%8F%AF%E6%A2%A6%E5%88%97%E8%A1%A8%EF%BC%88%E6%8C%89%E5%85%A8%E5%9B%BD%E5%9B%BE%E9%89%B4%E7%BC%96%E5%8F%B7%EF%BC%89"
sel = GetSelect(national_list, "td")
dic = {
    "number": "",
    "url": "",
    "cName":"",
    "jName":"",
    "eName":"",
    "attr1":"",
    "attr2":"",
    "attr1_url":"",
    "attr2_url":"",
    "hp":"",
    "atk":"",
    "def":"",
    "satk":"",
    "sdef":"",
    "speed":""
}
j={"data":[]}
selToList=[]
for i in sel:
    selToList.append(i)
# print(len(selToList))
print(main_url[:-4])
for i in range(len(selToList)):
    if selToList[i].string!=None:
        if "#" in selToList[i].string:
            print(selToList[i+1].a["href"])
            dic["number"] = selToList[i].string
            dic["cName"] = selToList[i+1].a["title"] if "title" in str(selToList[i+1].a) else None
            dic["jName"] = selToList[i+3].string
            dic["eName"] = selToList[i+4].string
            dic["url"] = selToList[i+1].a["href"] if "href" in str(selToList[i+1].a) else None
            dic["attr1"] = selToList[i+5].a["title"] if "title" in str(selToList[i+5].a) else None
            dic["attr1_url"] = selToList[i+5].a["href"] if "href" in str(selToList[i+5].a) else None
            dic["attr2"] = selToList[i+6].a["title"] if "title" in str(selToList[i+6].a) else None
            dic["attr2_url"] = selToList[i+6].a["href"] if "href" in str(selToList[i+6].a) else None
            getAttr=GetHp(main_url[:-4]+dic["url"])
            dic["hp"] = next(getAttr)
            dic["atk"] = next(getAttr)
            dic["def"] = next(getAttr)
            dic["satk"] = next(getAttr)
            dic["sdef"] = next(getAttr)
            dic["speed"] = next(getAttr)
            j["data"].append(dic.copy())
            dic.clear()
json1 = json.dumps(j)
with open("data.json","w") as outfile:
    json.dump(j,outfile)
```
